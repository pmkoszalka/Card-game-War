```python
import random
value = {'Two':2, 'Three':3, 'Four':4, 'Five':5,'Six':6,'Seven':7, 'Eight':8,'Nine':9,'Ten':10, 'Jack':11,'Queen':12,'King':13,'Ace':14}
suit = ('Hearts','Spades', 'Clubs', 'Diamonds')
rank = ('Two', 'Three', 'Four', 'Five','Six','Seven', 'Eight','Nine','Ten', 'Jack','Queen','King','Ace')
```

```python
class Card():
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = value[rank]
    def __str__(self):
        return f'{self.rank} of {self.suit}'
```

```python
class Deck():
    def __init__(self):
        self.all_cards=[]
        
        for x in suit:
            for y in rank:
                
                created_card=Card(x,y)
                
                #Create the card object:
                self.all_cards.append(created_card)
                
    def shuffle(self):
        random.shuffle(self.all_cards) # this does everything inplace to the list!!!!
    
    def deal_one(self):
        return self.all_cards.pop()
```

```python
class Player():
    def __init__(self,name):
        self.name=name
        self.all_cards=[]
        
    def remove_one(self):
        return self.all_cards.pop(0)
    
    def add_cards(self,new_cards):
        if type(new_cards)==type([]):
            return self.all_cards.extend(new_cards)
        else:
            return self.all_cards.append(new_cards)
    
    def __str__(self):
        return f'Player {self.name} has {len(self.all_cards)} cards.'
```

```python
game_on = True
at_war = False
cards_player_one = []
cards_player_two=[]

player_one=Player('Paweł')
player_two=Player('MareK')
new_deck = Deck()
new_deck.shuffle()
player_one.add_cards(new_deck.all_cards[0:26])
player_two.add_cards(new_deck.all_cards[26:])
counter = 1

while game_on:
    print(" ")
    while at_war:
        print('\033[1mWe are at war! \033[0m')
        for i in range(5):
            cards_player_one.append(player_one.remove_one()) 
            cards_player_two.append(player_two.remove_one())
            
        
        if cards_player_one[-1].value>cards_player_two[-1].value:
            player_one.add_cards(cards_player_one)
            player_one.add_cards(cards_player_two)
            cards_player_one = []
            cards_player_two=[]
            at_war= False
            break
        
        elif cards_player_one[-1].value<cards_player_two[-1].value:
            player_two.add_cards(cards_player_one)
            player_two.add_cards(cards_player_two)
            cards_player_one = []
            cards_player_two=[]
            at_war= False
            break
   # print(' ')
    print(f'\033[1mRound: {counter}\033[0m')
    counter +=1
    print(player_one)
    print(player_two)
    
    cards_player_one.append(player_one.remove_one())
    cards_player_two.append(player_two.remove_one())
    
    if cards_player_one[0].value>cards_player_two[0].value:
        player_one.add_cards(cards_player_one)
        player_one.add_cards(cards_player_two)
        cards_player_one = []
        cards_player_two=[]
        
    elif cards_player_one[0].value<cards_player_two[0].value:
        player_two.add_cards(cards_player_one)
        player_two.add_cards(cards_player_two)
        cards_player_one = []
        cards_player_two=[]
    else:
        if len(player_one.all_cards)<5:
            print('\033[1m'+ 'Player Two wins\033[0m')
            game_on = False
        elif len(player_two.all_cards)<5:
            print('\033[1m'+ 'Player One wins\033[0m')    
            game_on = False
    
        at_war = True
    
    if len(player_one.all_cards)<5:
        print('\033[1m'+ 'Player Two wins\033[0m')
        game_on = False
    elif len(player_two.all_cards)<5:
        print('\033[1m'+ 'Player One wins\033[0m')    
        game_on = False
```
